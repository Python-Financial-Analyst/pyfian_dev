# Floating Rates

- Introduce the FloatingRateNoteClass.

## Using the Floating Rate Class

```python
import pandas as pd
from pyfian.fixed_income.floating_rate_note import FloatingRateNote
import IPython.display as ipd

# Define a minimal floating rate note
frn = FloatingRateNote(
    issue_dt="2025-01-01",
    maturity="2030-01-01",
    ref_rate_curve=None,  # Optional, can be left None for minimal input
    quoted_margin=0.02,  # 2% over reference rate
    cpn_freq=2,  # Semi-annual coupons
    notional=1000,  # Face value
    settlement_date="2025-12-01",
)

# Build a table with minimum necessary inputs
inputs_data = [
    {"Attribute": "Instrument", "Value": frn.__class__.__name__},
    {"Attribute": "Issue Date", "Value": frn.issue_dt},
    {"Attribute": "Maturity Date", "Value": frn.maturity},
    {"Attribute": "Settlement Date", "Value": frn.get_settlement_date()},
    {"Attribute": "Notional", "Value": frn.notional},
    {"Attribute": "Coupon Frequency", "Value": frn.cpn_freq},
    {"Attribute": "Quoted Margin", "Value": frn.quoted_margin},
]

# Convert to DataFrame
df_inputs = pd.DataFrame(inputs_data)

# Display table without index
ipd.display(df_inputs.style.hide(axis="index"))
```
